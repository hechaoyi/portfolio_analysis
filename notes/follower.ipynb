```python
%pylab inline
from wallet.util.analysis import Analysis
```

```python
ana = Analysis.from_hedge_funds(126, 5, additions=['SCHG','ATVI','SHOP','DOCU'], max_cap=500, min_cap=5,
  categories={'Consumer Durables','Consumer Services','Retail Trade','Technology Services'})
print(ana)
corr = ana.data.rolling(5).mean().pct_change().corr().loc['SCHG']
print("correlation median:", corr.median())
ana.setup_mask(corr[corr <= corr.median()].index)
ana.screen()
```

```python
res = ana.optimize_iteration([100], max_count=2, additions=['ATVI','JD'])
print(sorted({s for _,d in res[0] for s in d.keys()}))
res
```
