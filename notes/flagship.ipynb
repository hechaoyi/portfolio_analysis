```python
%pylab inline
from wallet.util.analysis import Analysis
sectors = ['SCHX','SCHG','SCHV','SCHA','SCHF','SCHE',
           'CNYA','KBA','ASHR','GXC','MCHI','CQQQ','KWEB',
           'VGLT','SCHP','TVIX',]
```

```python
ana = Analysis(sectors, 21, 5)
print(ana)
ana.screen()
```

```python
ana.data.rolling(5).mean().pct_change().corr().loc['SCHG'].sort_values()
```

```python
ana.setup_mask(['SCHG','CNYA','VGLT','SCHP','TVIX'])
ana.optimize_iteration([100], min_percent=.1)
```

```python
ana.graph({'SCHG':45,'CNYA':15,'VGLT':30,'SCHP':8,'TVIX':2}, drop_components=['TVIX'])
```
